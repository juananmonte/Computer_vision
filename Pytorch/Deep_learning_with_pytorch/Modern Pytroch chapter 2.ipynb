## Building a small neural network with Pytorch

```python
x = [[1,2],[3,4],[5,6],[7,8]]
y = [[3],[7],[11],[15]]
```

```python
#1. Convert the input and output into tensors
X = torch.tensor(x).float()
Y = torch.tensor(y).float()
#It is good practice to have tensor objects as floats or long ints, as they will be multiplied by decimal values (weights) anyway
```

```python
#1.1 register the tensors to cuda
X = X.to(device)
Y = Y.to(device)
```

```python
#2. Define the Nerural Network Architecture

class MyNeuralNet(nn.Module):#It is mandatory to inherit from nn.Module when creating a model architecture as it is the base class for all neural network modules:
    def __init__(self): #initialize all the components of a Neural Network
        super().__init__() #so the class injerits the nn.Module functions
        
        #defining the components of the Neural network
        self.input_to_hidden_layer = nn.Linear(2,8) #in features (2) and out features (8)
        self.hidden_layer_activation = nn.ReLU()
        self.hidden_to_output_layer = nn.Linear(8,1)
        
    #Connect the components together and put them into a forward propagation function    
    def forward(self, x): #t is mandatory to use forward as the function name.
        x = self.input_to_hidden_layer(x)
        x = self.hidden_layer_activation(x)
        x = self.hidden_to_output_layer(x)
        return x
```

```python
#3. Generate randomized weights 

#just put the Neural Network on GPU first
mynet = MyNeuralNet().to(device)
```

```python
mynet.input_to_hidden_layer.weight

#These are random and will change everytime. If we want them not to change we use manual_seed method in Torch : 'torch.manual_seed(0)'
```

```python
# se the parameters of the model
for par in mynet.parameters():
 print(par)
```

```python
#4. Enough exploring. Define the loss function that we optimize for

loss_func = nn.MSELoss()

_Y = mynet(X)
loss_value = loss_func(_Y, Y) #The calculated values from the 'mynet' Neural Network (_Y) and the original output (Y)
print(loss_value)
```

```python
#5. Define the optimizer to reduce that loss. For this example we will use SGD
from torch.optim import SGD
opt = SGD(mynet.parameters(), lr=0.001)
```

```python
#6. Put everything together to run it on an epoch

#for that we follow thes steps

# 1. Calculate the loss value corresponding to the given input and output.
# 2.Calculate the gradient corresponding to each parameter.
# 3. Update the weights based on the learning rate and gradient of each parameter.
# 4. Once the weights are updated, ensure that the gradients that have been calculated in the previous step are flushed 
#    before calculating the radients in the next epoch:
```

```python
loss_history = []

for _ in range(50):
    opt.zero_grad() #step 4
    loss_value = loss_func(mynet(X), Y) #step1
    loss_value.backward() # step2
    opt.step() #step 3
    loss_history.append(loss_value.cpu().data.numpy()) #convert to cpu and then to numpy
```

```python
plt.plot(loss_history)
plt.title('Loss variation over increasing epochs')
plt.xlabel('epochs')
plt.ylabel('loss value')
```

## Dataset, DataLoader, and batch size

```python
x = [[1,2],[3,4],[5,6],[7,8]]
y = [[3],[7],[11],[15]]

X = torch.tensor(x).float()
Y = torch.tensor(y).float()

X = X.to(device)
Y = Y.to(device)
```

```python
#2. Define the class MyDataset. With this class we fetch one data point at a time and group them together (using DataLoader)
class MyDataset(Dataset):
    def __init__(self, x, y):
        self.x = torch.tensor(x).float()
        self.y = torch.tensor(y).float()
    
    #specify the len of the input dataset
    def __len__(self):
        return len(self.x)
    
    #Define the getitem function to fetch a specific row
    def __getitem__(self, ix):
        return self.x[ix], self.y[ix] #ix is the index of the row
```

```python
#3. Create an instance of the MyDataset class
ds = MyDataset(X, Y)
```

```python
dl = DataLoader(ds, batch_size=2, shuffle=True) #Dataloader is used so we can get batch sizes
```

```python
for x,y in dl:
 print(x,y)
```

```python
# 4. define the neural network again
class MyNeuralNet(nn.Module):
    def __init__(self):
        super().__init__()
         self.input_to_hidden_layer = nn.Linear(2,8)
         self.hidden_layer_activation = nn.ReLU()
         self.hidden_to_output_layer = nn.Linear(8,1)
    def forward(self, x):
        x = self.input_to_hidden_layer(x)
        x = self.hidden_layer_activation(x)
        x = self.hidden_to_output_layer(x)
        return x
```

```python
# 5. We get again the model to GPU, then the loss and optimizer
mynet = MyNeuralNet().to(device)
loss_func = nn.MSELoss()
from torch.optim import SGD
opt = SGD(mynet.parameters(), lr = 0.001)
```

```python
# 6. We loop through 50 epochs washing off the gradients,calculating the loss, making gradient descent, optimizing the values and appending the loss

import time
loss_history = []
start = time.time()
for _ in range(50):
    for data in dl:
         x, y = data
         opt.zero_grad()
         loss_value = loss_func(mynet(x),y)
         loss_value.backward()
         opt.step()
         loss_history.append(loss_value)
end = time.time()
print(end - start)
```
